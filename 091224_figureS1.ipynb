## old AuNR batch

```python
# plot UV_vis spectra
p_data = Path("C:/Users/katsuya2/OneDrive - University of Illinois - Urbana/Documents/Python Scripts/projects/AuNR_DA/datasets")
uvvis_data = pd.read_csv(Path(p_data, 'uvvis_AuNR25x85.csv'))
TEM_data = pd.read_csv(Path(p_data, 'tem_AuNR25x85.csv'))
```

```python
# Create random data for demonstration
width = TEM_data['Minor']
length = TEM_data['Major']

# Create a figure with an appropriate size
fig = plt.figure(figsize=(10, 15))

# Create subplots using subplot2grid
# Top row plot (A)
axA = plt.subplot2grid((3, 2), (0, 0), colspan=2)  # Span across the first row
axA.plot(uvvis_data['wav'], uvvis_data['absorbance'], c='k', lw=4)
axA.text(-0.13, 1, 'A', ha='left', va='top', transform=axA.transAxes, fontsize=30)
axA.tick_params(axis='both', labelsize=30, pad=3)
axA.set_xlim(400, 800)
axA.set_xticks([500, 600, 700])
axA.set_xlabel('Wavelength (nm)', fontsize=30)
axA.set_ylabel('Absorbance', fontsize=30)

# Second row, first column (B)
axB = plt.subplot2grid((3, 2), (1, 0))
axB.text(-0.23, 1, 'B', ha='left', va='top', transform=axB.transAxes, fontsize=30)
axB.set_xticks([])
axB.set_yticks([])

# Third row, first column (C)
axC = plt.subplot2grid((3, 2), (2, 0))
binwidth = 2
target = length
mean = np.mean(target)
std = np.std(target)
target_divided_mean = target/mean
axC.hist(target, bins=np.arange(min(target), max(target)+binwidth, binwidth), 
        color='r', label=f'$\mu$ = {int(round(mean, 0))}, $\sigma$ = {int(round(std, 0))}')
axC.text(-0.23, 1, 'C', ha='left', va='top', transform=axC.transAxes, fontsize=30)
axC.tick_params(axis='both', labelsize=30, pad=3)
axC.set_xlabel('Length (nm)', fontsize=30)
axC.set_ylabel('Occurrences', fontsize=30)
axC.set_ylim(0, 32)
axC.set_xlim(50, 120)
axC.set_xticks([60.0, 80.0, 100.0])
axC.set_yticks([0, 10, 20, 30])
axC.legend(fontsize=20)

# Second row, second column (D)
axD = plt.subplot2grid((3, 2), (1, 1))
binwidth = 0.5
target = width
mean = np.mean(target)
std = np.std(target)
axD.hist(target, bins=np.arange(min(target), max(target)+binwidth, binwidth), 
        color='b', label=f'$\mu$ = {int(round(mean, 0))}, $\sigma$ = {int(round(std, 0))}')
axD.text(-0.23, 1, 'D', ha='left', va='top', transform=axD.transAxes, fontsize=30)
axD.tick_params(axis='both', labelsize=30, pad=3)
axD.set_xlabel('Width (nm)', fontsize=30)
axD.set_ylabel('Occurrences', fontsize=30)
axD.set_ylim(0, 60)
axD.set_xticks([26.0, 30.0, 34.0])
axD.set_yticks([0, 20, 40])
axD.legend(fontsize=20)

# Third row, second column (E)
axE = plt.subplot2grid((3, 2), (2, 1))
binwidth = 0.1
target = length/width
mean = np.mean(target)
std = np.std(target)
axE.hist(target, bins=np.arange(min(target), max(target)+binwidth, binwidth), 
        color='g', label=f'$\mu$ = {round(mean, 1)}, $\sigma$ = {round(std, 1)}')
axE.text(-0.23, 1, 'E', ha='left', va='top', transform=axE.transAxes, fontsize=30)
axE.tick_params(axis='both', labelsize=30, pad=3)
axE.set_xlabel('Aspect Ratio', fontsize=30)
axE.set_ylabel('Occurrences', fontsize=30)
axE.set_ylim(0, 32)
axE.set_xlim(2, 4.5)
axE.set_xticks([2.5, 3.5])
axE.set_yticks([0, 10, 20, 30])
axE.legend(fontsize=20)

# Adjust layout to prevent overlap
fig.tight_layout()
```
